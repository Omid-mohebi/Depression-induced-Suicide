```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
train_data
```

```python
test_data = pd.read_csv('test.csv')
test_data
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
stop_words
```

```python
X_train = (train_data['NarrativeLE']+train_data['NarrativeCME']).to_numpy()
X_test = (test_data['NarrativeLE']+test_data['NarrativeCME']).to_numpy()
Y_train = train_data['DepressedMood'].to_numpy()
Y_train = np.array([1 if y == 1 else 0 for y in Y_train]).reshape(-1, 1)
X_train = [txt.split() for txt in X_train]
X_test = [txt.split() for txt in X_test]
```

```python
def tokenize(data):
    temp = []
    for d in data:
        if len(d)>3 and d not in stop_words:
            temp.append(d)
    return temp
```

```python
X_train = [tokenize(text) for text in X_train]
X_test = [tokenize(text) for text in X_test]
```

```python
wordtoidx = {}
idx = 0
for text in X_train:
    for txt in text:
        if txt not in wordtoidx:
            wordtoidx[txt] = idx
            idx += 1
for text in X_test:
    for txt in text:
        if txt not in wordtoidx:
            wordtoidx[txt] = idx
            idx += 1
len(wordtoidx)
```

```python
D = len(wordtoidx)
N = len(X_train)
N2 = len(X_test)
# M = 10
K = 1
```

```python
train_vec = np.zeros((N, D))
test_vec = np.zeros((N2, D))
i = 0
for row in X_train:
    for text in row:
        train_vec[i, wordtoidx[text]] += 1
    i += 1
i = 0
for row in X_test:
    for text in row:
        test_vec[i, wordtoidx[text]] += 1
    i += 1

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(train_vec, Y_train)
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
py = model.predict(test_vec)
py
```

```python
# evaluate model
model.score(xtest, ytest) * 100
```

```python
# predict test samples
submission = pd.DataFrame({'DepressedMood': py})
submission
```
